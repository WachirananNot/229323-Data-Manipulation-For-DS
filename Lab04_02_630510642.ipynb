```python
import pandas as pd
```

# 2.

```python
file = open("found_pokemon.txt", "r", encoding="utf-8")
data = []
head_line = []
first = True
for line in file:
  line = line.rstrip().split(",")
  if first:
    head_line = line
    first = False
  else:
    line[1] = int(line[1])
    line[2] = int(line[2])
    line[3] = int(line[3])
    data.append(tuple(line))

found = pd.DataFrame(data=data,index=['']*len(data),columns = head_line)
found
```

# a)

```python
found["sum found"] = found["market"]+found["playground"]+found["shopping mall"]
found
```

# b)

```python
market = [found['kind'][x] for x in range(len(found['market'])) if found['market'][x] == found['market'].max()]
playground =[found['kind'][x] for x in range(len(found['playground'])) if found['playground'][x] == found['playground'].max()]
shopping = [found['kind'][x] for x in range(len(found['shopping mall'])) if found['shopping mall'][x] == found['shopping mall'].max()]
place = ['market','playground','shopping mall']
kind = []
kind.extend(market)
kind.extend(playground)
kind.extend(shopping)
max_place = pd.DataFrame(data=kind,index=place,columns=[''])
max_place
```

# c)

```python
found["per_market_x"] = (found["market"]/found['sum found'])*100
found["per_playground_x"] = (found["playground"]/found['sum found'])*100
found["per_shopping_x"] = (found["shopping mall"]/found['sum found'])*100
found
```

# d)

```python
top_3_mar = found.sort_values(by=["market"],ascending = False)
top_3_mar = top_3_mar.iloc[:, [0,1]].head(3)
top_3_play = found.sort_values(by=["playground"],ascending = False)
top_3_play = top_3_play.iloc[:, [0,2]].head(3)
top_3_shop = found.sort_values(by=["shopping mall"],ascending = False)
top_3_shop = top_3_shop.iloc[:, [0,3]].head(3)
top_3_place = pd.concat([top_3_mar,top_3_play,top_3_shop], axis=0)
top_3_place.fillna('-')
```
