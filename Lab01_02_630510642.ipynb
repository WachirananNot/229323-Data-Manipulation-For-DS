```python
import numpy as np
```

```python
a = int(input("Input a: "))
b = int(input("Input b: "))
m = int(input("Input m: "))
n = int(input("Input n: "))
array_1 = (np.arange(a*b)).reshape(a,b)
array_2 = (np.arange(m*n)).reshape(m,n)
print(array_1)
print(array_2)
add(array_1,array_2)
multi(array_1,array_2)
mulk(array_1)



```

```python
def add(array_1,array_2):
  try:
    array_3 = array_1+array_2
    print("result =",array_3)
  except ValueError:
    print("This calculation is wrong.")
```

```python
def multi(array_1,array_2):
  try:
    array_3 = np.dot(array_1,array_2)
    print("result =",array_3)
  except ValueError:
    print("This calculation is wrong.")
```

```python
def mulk(array_1):
  k = int(input("Input k: "))
  try:
    array_3 = np.dot(array_1,k)
    print("result =",array_3)
  except ValueError:
    print("This calculation is wrong.")
```
