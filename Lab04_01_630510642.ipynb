```python
import pandas as pd
```

```python
name = ['Kind','Found','Catch']
mon = pd.read_csv("doremon_data.csv",names=name,header=None)
mee = pd.read_csv("doremee_data.csv",names=name,header=None)
```

### 1.

# a)

```python
mon['Performance'] = ((mon['Catch']/mon['Found'])*100)
mon
```

```python
mee['Performance'] = ((mee['Catch']/mee['Found'])*100)
mee
```

# b)

```python
mon = mon.rename(columns={"Found":"mon_found","Catch":"mon_caught"})
mon
```

```python
mee = mee.rename(columns={"Found":"mee_found","Catch":"mee_caught"})
mee
```

```python
mon = mon.drop(labels = "Performance",axis = 1)
mee = mee.drop(labels = "Performance",axis = 1)
```

```python
all = pd.merge(mon,mee,how='outer')
all.fillna(0,inplace=True)
all["sum_caught"] = all["mon_caught"]+all["mee_caught"]
all["sum_found"] = all["mon_found"] + all["mee_found"]
all
```

# c)

```python
all["performance"] = (all["sum_caught"] / all["sum_found"])*100
all
```

# d)

```python
all.sort_values(by=['Kind'], inplace=True, ascending=False)
all.reset_index(drop=True, inplace=True)
all
```

# e)

```python
meemon = all.loc[(all["mee_caught"] > 0.0) & (all["mon_caught"] == 0.0), ["Kind"]]
meemon.sort_values(by=['Kind'],inplace=True)
meemon
```

# f)

```python
fo_ca = pd.concat([all["Kind"],all["sum_found"],all["sum_caught"]],axis = 1, join="inner")
fo_ca.sort_values(by=["sum_found","sum_caught"],ascending = [False, False],inplace=True)
fo_ca.reset_index(drop=True, inplace=True)
fo_ca
```

# g)

```python
perform = all.loc[(all["performance"] >= 65)]
perform
```

# h)

```python
max_min = pd.concat([all.iloc[[all["performance"].idxmax()]],all.iloc[[all["performance"].idxmin()]]])
max_min
```
