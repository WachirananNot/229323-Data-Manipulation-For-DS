```python
import numpy as np
from google.colab import files
import csv
```

```python
def find_head():
  files = open("630510642_01.csv","r")
  data_list = []
  for line in files:
    x = line.strip().split(',')
    data_list.append(x)
  head = data_list[0]
  del data_list[0]
  head.append("BMI")
  head.append("Result")
  return head
```

```python
#variable_name = files.upload()
head = find_head()
files = open("630510642_01.csv","r")
data_list = []
header = []

data_list = []
first = True
for line in files:
  if first:
    header = line
    first = False
  else:
    x = line.rstrip().split(',')
    x[1],x[2] = float(x[1]),float(x[2])
    data_list.append(x)
data_list = BMI_calculate(data_list)

print(data_list)
```

```python
def BMI_calculate(value):
  for i in value:
    BMI = i[2]/(i[1]**2)
    i.append(round(BMI,4))
    if BMI<=18.5:
      i.append("Thin")
    elif 18.6<=BMI<=24.9:
      i.append("Healthy")
    elif 25<=BMI<=29.9:
      i.append("Overweight")
    else:
      i.append("Obese")
  return value 
```

```python
csvfile = open("630510642_02.csv","w",newline='')
obj = csv.writer(csvfile)
obj.writerow(head)
for i in data_list:
  obj.writerow(i)

csvfile.close()
```
