```python
files = open("person_data_01.txt","r",encoding="utf-8")
header = []
data_list = []
first = True
for line in files:
  if first:
    header = line
    first = False
  else:
    x = line.rstrip().split(',')
    x[1],x[2] = float(x[1]),float(x[2])
    data_list.append(tuple(x))

height = []
weight = []
for i in data_list:
  height.append(i[1])
  weight.append(i[2])
height = np.array(height)
weight = np.array(weight)
print("b)")
print("Mean height:",np.mean(height))
print("Median height:",np.median(height))
print("Mode height:",stats.mode(sorted(height)))
print()
print("Mean weight:",np.mean(weight))
print("Median weight:",np.median(weight))
print("Mode weight:",stats.mode(sorted(weight)))
```

```python
csvfile = open("630510642_01.csv","w",newline='')
obj = csv.writer(csvfile)
obj.writerow(head)
for i in data_list:
  obj.writerow(i)

csvfile.close()
```
