```python
import numpy as np
from google.colab import files
import csv
import pandas as pd
from tabulate import tabulate
```

```python
# variable_name = files.upload()
files = open("stu_grade.txt","r")
list_ = files.read()

list_ = list_.replace(" =",":").split("\n")
new_list = []
for i in range (len(list_)):
   list_[i] = list_[i].split(":")
   try:
      list_[i][1] = list_[i][1].strip()
   except IndexError:
      continue

for i in range(len(list_)):
  x = []
  
  try:
    x.append(list_[i][1])
  except IndexError:
    break
  new_list.append(x)
new_list2 = []
k = 0
x = []
for i in range(len(new_list)):
  x.extend(new_list[k])
  k+=1
  if k %3 == 0:
    new_list2.append(x)
    x = []

py=[]
java = []
c = []
lower_ = []
higher_ = []
for i in new_list2:
  if i[1] == 'Python':
    py.append(i[0])
  elif i[1] == 'C':
    c.append(i[0])
  elif i[1] == 'Java':
    java.append(i[0])
  if float(i[2])>= 3.00:
    higher_.append(i[0])
  elif 2.00<= float(i[2])<=2.99:
    lower_.append(i[0])
```

```python
print("1)")
Python = {}
Java = {}
C = {}
Python[len(py)] = py
Java[len(java)] = java
C[len(c)] = c
P = pd.DataFrame(data=Python,index=[""]*len(py))
J = pd.DataFrame(data=Java,index=[""]*len(java))
Cc = pd.DataFrame(data=C,index=[""]*len(c))
P.columns.name = 'Python'
J.columns.name = "Java"
Cc.columns.name = "C"
print(P)
print(J)
print(Cc)
```

```python
print("2)")
two = pd.DataFrame(lower_,index = [""]*len(lower_))
three = pd.DataFrame(higher_,index = [""]*len(higher_))
two.columns = [''] * len(two.columns)
three.columns = [''] * len(three.columns)
two.columns.name = "2.00-2.99"
three.columns.name = ">= 3.00"
two.append(three)
print(two.to_string())
print(three.to_string())
```
