```python
import numpy as np
```

```python
n = int(input("Enter n :"))
array_data = (np.arange(n*n)).reshape(n,n)
print("Sum row is :",find_sum_row(array_data))
print("Sum col is :",find_sum_col(array_data))
print("Sum diagonal is :",find_sum_diagonal(array_data))
print("Sum total is :",array_data.sum())
print("Sum transpose is :",sum_transpose(array_data))
```

```python
def find_sum_row(array_data):
  sum_row = np.zeros(n,dtype=np.int32)
  for i in range(n):
    for j in range(n):
      sum_row[i] = array_data[i][:].sum()
  return sum_row
```

```python
def find_sum_col(array_data):
  sum_col = np.zeros(n,dtype=np.int32)
  for i in range(n):
    for j in range(n):
      sum_col[i] = array_data[:,i].sum()
  return sum_col
```

```python
def find_sum_diagonal(array_data):
  sum_diagonal = np.zeros(2,dtype=np.int32)
  sum_diagonal[0] = (np.diag(array_data)).sum()
  sum_diagonal[1] = (np.diag(np.fliplr(array_data))).sum()
  return sum_diagonal
```

```python
def sum_transpose(array_data):
  T = array_data.T
  sum_transpose = np.add(array_data,T)
  return sum_transpose
```
